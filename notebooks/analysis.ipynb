```python
# -------------------------------
# Creazione dataset simulato guasti segnalamento ferroviario
# -------------------------------

import pandas as pd
import numpy as np

# numero di eventi simulati
n = 1000

data = {
    "date": pd.date_range(start="2024-01-01", periods=n, freq="H"),
    
    "line": np.random.choice(
        ["Line_A", "Line_B", "Line_C"], size=n
    ),
    
    "station": np.random.choice(
        ["Genova Brignole", "Genova Nervi", "Milano Centrale", 
         "Milano Rogoredo", "Torino Porta Nuova", "Torino Porta Susa"], size=n
    ),
    
    "system_type": np.random.choice(
        ["ACC", "Electronic Relay", "Cabin Signaling"], size=n
    ),
    
    "failure_type": np.random.choice(
        ["Signal Loss", "Communication Error", "Power Failure", "Hardware Fault"], size=n
    ),
    
    "intervention_minutes": np.random.randint(5, 180, size=n)
}

df_simulated = pd.DataFrame(data)

# salvataggio CSV
df_simulated.to_csv("../data/signaling_failures.csv", index=False)

print("Dataset creato correttamente in data/signaling_failures.csv")
```
